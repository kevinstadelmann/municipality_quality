### Calculation of longitude and latitude

### Approximation formula:

1. Form auxiliary variables

$$ y' = \frac{E - 2600000 m}{1000000}\ = \frac{y - 600000 m}{1000000}\ $$

$$ x' = \frac{N - 1200000 m}{1000000}\ = \frac{x - 200000 m}{1000000}\ $$

2. Length $\lambda$ and width $\phi$ calculate in the unit [10000"]

$$ \lambda ' = 2.6779094 + 4.728982 * y' + 0.791484 * y' * x' + 0.1306 * y' * x'^2 - 0.0436 * y'^3 $$
                                                            
$$ \phi ' = 16.9023892 + 3.238272 * x' - 0.270978 * y'^2 - 0.002528 * x'^2 - 0.0447 * y'^2 * x' - 0.0140 * x'^3 $$
                                                            
3. Convert the length and width to the unit

$$ \lambda =  \frac{\lambda ' * 100}{36} $$
                    
$$ \phi =  \frac{\phi ' * 100}{36} $$


```python
munCoord['y_aux'] = (munCoord['E_CNTR'] - 2600000) / 1000000
munCoord['x_aux'] = (munCoord['N_CNTR'] - 1200000) / 1000000

munCoord['lat'] = (16.9023892 \
        +  3.238272 * munCoord['x_aux'] \
        -  0.270978 * munCoord['y_aux']**2 \
        -  0.002528 * munCoord['x_aux']**2 \
        -  0.0447   * munCoord['y_aux']**2 * munCoord['x_aux'] \
        -  0.0140   * munCoord['x_aux']**3)
munCoord['lat'] = round(munCoord['lat'] * 100/36, 5)

munCoord['lng'] = (2.6779094 
        + 4.728982 * munCoord['y_aux']
        + 0.791484 * munCoord['y_aux'] * munCoord['x_aux']
        + 0.1306 * munCoord['y_aux'] * munCoord['x_aux']**2 
        - 0.0436 * munCoord['y_aux']**3)
munCoord['lng'] = round(munCoord['lng'] * 100/36, 5)
```
